```python
sns.histplot(df['Carat Weight'], kde=True)
plt.title('Распределение Carat Weight')
plt.xlabel('Carat Weight')
plt.ylabel('Частота')
plt.show()
```

Распределение веса карата у бриллиантов. Самое частое значение это 1 карат. Большая часть бриллиантов с каратом меньше 1.5. Меньше 25% бриллиантов с каратом больше 2.0, чаще всего в данном дмапазоне встречается 2 карата. Это означает что большинство бриллиантов малого размера. Самый редкий ответ это приблизительно 2.8 карата.

```python
sns.countplot(x='Color', data=df)
plt.title('Распределение Color')
plt.xlabel('Color')
plt.ylabel('Количество')
plt.show()
```

Классификация цвета у бриллиантов от чистого к грязному: D, E, F, G, H, I. Больше всего бриллиантов с цветом G, что означает не самый чистый цвет, но и не самый худший. Меньше всего с цветом D, а по классификации это чистейший цвет у бриллиантов, что означает качественных бриллиантов меньше всего.

```python
sns.boxplot(x=df['Price'])
plt.title('Boxplot Price')
plt.xlabel('Price')
plt.show()
```
